# Python `sort()` 정렬 알고리즘
파이썬의 `list.sort()`와 `sorted()` 함수는 **Timsort(팀정렬)** 알고리즘을 사용합니다.

## 📌 Timsort란?
- Python과 Java에서 사용되는 **하이브리드 정렬 알고리즘**
- **병합 정렬(Merge Sort)** + **삽입 정렬(Insertion Sort)** 기반
- 실제 데이터에서 자주 나타나는 **부분적으로 정렬된 패턴(run)** 을 매우 효율적으로 처리하도록 설계됨

## ✔ 특징
- **O(n log n)** 시간 복잡도 (평균 / 최악)
- 최선의 경우 **O(n)** (이미 정렬된 경우)
- 안정 정렬(Stable Sort)
- 실사용 데이터에서 성능이 매우 좋음

## ✔ 코드 예시
아래는 Timsort가 동작하는 기본 코드 예시입니다.

```python
data = [5, 3, 1, 4, 2]
data.sort()  # Timsort 실행
data
```
