```python
g={
    1:[2,3],
    2:[1,4,5],
    3:[1],
    4:[2],
    5:[2]
}
```

```python
def bfs(graph, start):
    que = []
    done = set()
    que.append(start)
    done.add(start)
    while que:
        v = que.pop(0)
        print(v, end=" ")
        for nv in graph[v]:
            if nv not in done:
                que.append(nv)
                done.add(nv)
```

```python
bfs(g,1)
```
